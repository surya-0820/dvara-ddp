```r
hid = 1

household_df = read.csv(paste("D:\\DDP\\household_id_1.csv",sep=""),stringsAsFactors = FALSE)

household_df$closeqtr[is.na(household_df$closeqtr)] = '201904'
household_df$closeqtr[trimws(household_df$closeqtr,which='both')==''] = '201904'
household_df$closedt[is.na(household_df$closedt)] = '31-12-2019' 
household_df$closedt[trimws(household_df$closedt,which='both')==''] = '31-12-2019'
# unique loans
vec = household_df[c('household_id','contributortype','accttype','disburseddt','closedt','disbursedqtr','closeqtr','accountstatus')]
dt = as.data.table(vec)[, list(list(.I)), by = vec]
dd = as.data.frame.matrix(dt)
dd = within(dd, rm(V1))
```

```r
apply(dt, 1, function(dt_row){
    index = unlist(dt_row[9])[1]
    
    
    
    reported_dates_vector <- strsplit(toString(household_df$reporteddatehist[index]),',')[[1]]
    
    
    unique_indices = seq_along(reported_dates_vector)[!duplicated(reported_dates_vector)] ## Unique reported dates - returns indices from 1 to 36
    reported_dates_vector = reported_dates_vector[unique_indices]    
    dpdhist_vector <- household_df[index, paste('dpdhist',unique_indices,sep="",collaspe="")] ##unique_indices
    #print(dpdhist_vector)
    dpdhist_vector[is.na(dpdhist_vector)] = 0
    dpdhist_vector <- as.numeric(dpdhist_vector)
    amountoverdue <- household_df$amtoverduehist[index]
    curbalhist <- household_df$curbalhist[index]
    quarter_list = sapply(reported_dates_vector, function(reported_dates_vector) 
      paste(substr(reported_dates_vector,start=1,stop=4), 
            switch(substr(reported_dates_vector,start=5,stop=6),"01"='01',"02"='01',"03"='01',"04"='02',"05"='02',"06"='02',"07"='03',"08"='03',"09"='03',"10"='04',"11"='04',"12"='04')
            ,sep="",collapse="")
    )
    unique_quarter_list = unique(quarter_list)
    
    
    # Observation:
    # unique quarter list is of length 12 compared to length 36 in quarter list
    # so there are on average 3 reported dates per quarter
    
    
    # find the max date from the reported dates belonging to the same quarter
    unique_dpdhist <- sapply(unique_quarter_list, function(quarter){
      a = which(quarter_list==quarter)  ##Indices for quarters for every unique quarter
      d = dpdhist_vector[a[which.max(reported_dates_vector[a])]]
      # print(paste('ddddddddddddddddddd::::::::;',d))
      return(d)
    })
    
    
    
    for(q in seq(1,nrow(household_info))){
      # print('in')
      if(!(household_info$loan_period[q] %in% unique_quarter_list)){next}
      i = which(unique_quarter_list == household_info$loan_period[q])
      household_info$dpd[q] <<- strtoi(household_info$dpd[q]) + strtoi(unique_dpdhist[i])
    }

})
```

```r
debt_service_df = household_info[,c(1,2,9,10)]  ##Taking household_id, loan_period, dpd, delinquency status 
income_vector <- rep(0,nrow(debt_service_df))
expense_vector <- rep(0,nrow(debt_service_df))

for(subrow in seq(1,nrow(household_df))){
p = strsplit(toString(household_df$period_ending[subrow]),'-')
p_yr = substr(p[[1]][1],start=2,stop=5)
p_qtr = switch(p[[1]][2],"01"=1,"02"=1,"03"=1,"04"=2,"05"=2,"06"=2,"07"=3,"08"=3,"09"=3,"10"=4,"11"=4,"12"=4)

period_ending_val = paste(toString(p_yr),'0',toString(p_qtr),sep="",collapse="")

income_vector[which(debt_service_df$loan_period %in% d[which(d == period_ending_val):length(d)])] = household_df[subrow,paste('income_quarter_',p_yr,'_',p_qtr,sep="")]
expense_vector[which(debt_service_df$loan_period %in% d[which(d == period_ending_val):length(d)])] = household_df[subrow,paste('total_expense_quarter_',p_yr,'_',p_qtr,sep="")]
}

debt_service_df = cbind(debt_service_df,income_annual=income_vector)
debt_service_df = cbind(debt_service_df,expense_quarter=expense_vector)

 
  # emi_df = read.csv(paste('EMI DF/emi_df_hid_',hid,'.csv',sep="",collaspe=""))
  emi_column = rep(0,nrow(debt_service_df))
  
  for(l in seq(1,nrow(emi_df))){  
    
    # account_in = which(npa_accstatus$disbursed_date == emi_df$disburseddt && npa_accstatus$close_date == emi_df$closedt)
    
    emi_year = strsplit(toString(emi_df$disburseddt[l]),'-')[[1]][3]
    emi_month = strsplit(toString(emi_df$disburseddt[l]),'-')[[1]][2]
    emi_qtr = switch(emi_month,"01"="01","02"="01","03"="01","04"="02","05"="02","06"="02","07"="03","08"="03","09"="03","10"="04","11"="04","12"="04")
    emi_dis_loan_period = paste(emi_year,emi_qtr,sep="",collapse="")  
    
    emi_year = strsplit(toString(emi_df$closedt[l]),'-')[[1]][3]
    emi_month = strsplit(toString(emi_df$closedt[l]),'-')[[1]][2]
    emi_qtr = switch(emi_month,"01"="01","02"="01","03"="01","04"="02","05"="02","06"="02","07"="03","08"="03","09"="03","10"="04","11"="04","12"="04")
    emi_close_loan_period = paste(emi_year,emi_qtr,sep="",collapse="")  
    
    if(emi_close_loan_period == "NA"){ emi_close_loan_period = "201904" }
    if(str_trim(toString(emi_df$closedt[1]),side='both') =='') {emi_close_loan_period = "201904"}
    emi_qtrs_period = d[which(d==emi_dis_loan_period):which(d==emi_close_loan_period)]
    emi_ind = which(debt_service_df$loan_period %in% emi_qtrs_period)
    
    for(e_in in emi_ind){
      loan_year = substr(debt_service_df$loan_period[e_in],start = 1, stop = 4)
      if(loan_year=='2019'){
        next
      }
      else{
        lyear = substr(debt_service_df$loan_period[e_in],start = 1, stop = 4)
        lqtr = substr(debt_service_df$loan_period[e_in],start = 6, stop = 6)
        val_e = emi_df[l,paste('emi_values.instmt_quarter_',lyear,'_',lqtr,sep="",collapse="")]
        if(is.null(val_e)){val_e = 0}
        else if(is.na(val_e)){val_e = 0}
        emi_column[e_in] = emi_column[e_in] + val_e
      }
    }
    # print('*******************************')
    # print(emi_column)
    # print(is.na(emi_column))
  }
  
  debt_service_ratio = rep(0,nrow(debt_service_df))
  for(e_in in seq(1,nrow(debt_service_df))){
    if(is.na(emi_column[e_in])) {debt_service_ratio[e_in] = 0}
    else if(emi_column[e_in] == 0){debt_service_ratio[e_in] = 0}
    else{ debt_service_ratio[e_in] = ((debt_service_df$income_annual[e_in]/4)-(debt_service_df$expense_quarter[e_in]/4))/emi_column[e_in] }
  }
  
  debt_service_df = cbind(debt_service_df, emi = emi_column)
  debt_service_df = cbind(debt_service_df, debt_service_ratio = debt_service_ratio)
```

```r
#Calculating previous quarter outstanding loans
outstanding_prev_quarter <- vector()
invisible(apply(household_info, 1, function(h_entry_row){
prev_qtr = d[which(d==h_entry_row[2])-1]
if(!(prev_qtr %in% household_info$loan_period)){
  outstanding_prev_quarter <<- c(outstanding_prev_quarter, 0)
}
else{
  ind = which(household_info$loan_period==prev_qtr)
  outstanding_prev_quarter <<- c(outstanding_prev_quarter, household_info$outstanding_loans_count[ind])
}
}))
# print(paste('Outstanding loans count previous quarter: ',outstanding_prev_quarter))

household_info = cbind(household_info, outstanding_loans_beginning  = outstanding_prev_quarter)
names(household_info)[names(household_info) == 'outstanding_loans_count'] <- 'outstanding_loans_end'
```

```r
##Calculating previous 5 quarters churn
  
  churn_prev_five_max = vector()

  invisible(apply(household_info, 1, function(h_entry_row){
    previous_values <- vector()
    # print(h_entry_row)
    max_churn = 0
    for(i in seq(1,5)){
      qtr_new = d[which(d==h_entry_row[2])-i]
      if(!(qtr_new %in% household_info$loan_period)){
        next
      }
      else{
        ind = which(household_info$loan_period==qtr_new)
        # print(max_churn)
        if(strtoi(household_info$churn[ind])>max_churn){
          max_churn = strtoi(household_info$churn[ind])
        }
      }
    }
    churn_prev_five_max <<- c(churn_prev_five_max, max_churn)
    # print(max_churn)
  }))
  
  household_info = cbind(household_info, churn_prev_five_max = churn_prev_five_max)
```

```r
  outstanding_loans_delta = vector() 
  debt_service_delta = vector()
  invisible(apply(household_info, 1, function(h_entry_row){
    prev_qtr = d[which(d==h_entry_row[2])-1]
    o=0
    ds=0
    if(!(prev_qtr %in% household_info$loan_period)){
      o=0
      ds=0
    }
    else{
      ind = which(household_info$loan_period==prev_qtr)
      o = household_info$outstanding_loans_beginning[ind]
      ds = household_info$debt_service_ratio[ind]
    }
    
    outstanding_loans_delta <<- c(outstanding_loans_delta,round(strtoi(h_entry_row[11])-o, digits = 3))  ##Current quarter outstanding - previous quarter outstanding
    debt_service_delta <<- c(debt_service_delta,round(as.numeric(h_entry_row[20])-ds, digits = 3))  ##Current quarter debt servie - previous quarter debt service
    
  }))
  
  household_info = cbind(household_info, outstanding_loans_delta = outstanding_loans_delta)
  household_info = cbind(household_info, debt_service_delta = debt_service_delta)
```
