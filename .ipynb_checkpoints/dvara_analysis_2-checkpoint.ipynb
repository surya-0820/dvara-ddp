```python
import pandas as pd
import numpy as np
import timeit
from tqdm import tqdm
from matplotlib import pyplot as plt
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
df1 = pd.read_csv("D:\DDP\preprocessed data\preprocessed_chkpt_0-100000.csv", index_col=0)
df2 = pd.read_csv("D:\DDP\preprocessed data\preprocessed_chkpt_100000-200000.csv", index_col=0)
df3 = pd.read_csv("D:\DDP\preprocessed data\preprocessed_chkpt_200000-300000.csv", index_col=0)
df4 = pd.read_csv("D:\DDP\preprocessed data\preprocessed_chkpt_300000-400000.csv", index_col=0)
df5 = pd.read_csv("D:\DDP\preprocessed data\preprocessed_chkpt_400000-494696.csv", index_col=0)

df = pd.concat([df1, df2, df3, df4, df5])
df = df.drop_duplicates()
df = df.reset_index(drop=True)
```

```python
df['year'] = df['loan_period'].apply(lambda x: x[:4])
q_list = np.sort(df.loan_period.unique()) # just a list from 2010Q1 to 2023Q1
```

Let's look at the distribution of quarter positions (w.r.t household rows) for rows with dpd=1e9

```python
df_1e9 = df[df.dpd==1e9]
quarter_pos_list = []

for row in tqdm(df_1e9.iterrows()): # iterate through rows that have missing dpd
    df_curr = df[df.household_id == row[1].household_id].reset_index() 
    pos = np.where(df_curr.loan_period == row[1].loan_period)[0][0]
    quarter_pos_list.append(pos)
    
plt.figure(figsize=(5, 4))
data = pd.Series(quarter_pos_list).value_counts().sort_index()
plt.bar(data.index, data.values)
plt.xlabel('Position of quarter (w.r.t rows for each household)')
plt.ylabel('Number of rows')
plt.title('Distribution of rows in terms of quarter positions \n (for rows with missing DPD)')
plt.show()

df_1e9['quarter_pos'] = quarter_pos_list # position of quarter in the original dataset
```

Rows with missing DPD occur mostly in 1st position w.r.t household. We can fill them with 0s.This makes sense as loan range period of households is not 1 quarter for any household.

```python
first_pos_hid = df_1e9[(df_1e9.quarter_pos == 0)].household_id # households with first quarter having missing dpd

df.loc[df.household_id.isin(first_pos_hid) & (df.dpd == 1e9), 'dpd'] = 0
```

```python
hid_list = df.household_id.unique() # unique household_ids, order preserved

ll = []

for hid in tqdm(hid_list):
    ll.extend(range(len(df[df.household_id == hid])))
df['quarter_pos'] = ll
```

```python
df['deliquency'] = df.dpd.apply(lambda x: x>5)
```
